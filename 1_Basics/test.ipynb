```python
import pandas as pd
```

```python
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
job_count = df.groupby("job_title_short")["salary_year_avg"].median().sort_values(ascending=True)


```

```python
job_count.plot(kind="barh")
plt.ylabel("Type of Job")
plt.xlabel("No. of Jobs")
plt.title("No. of Jobs by Title")
plt.show()
```
