```python
job_roles = "Data"
dfj = df[(df["job_title_short"].str.contains("Data", na=False)) & (df["job_country"] == "United States")]
dfj = dfj.dropna(subset="salary_year_avg")
dfj = dfj[dfj["salary_year_avg"] > 0]


sns.kdeplot(dfj["salary_year_avg"], fill=True, cmap="Blues", bw_adjust=0.7)

ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${int(x/1000)}K"))

plt.title("Salary Density of Data Related Job Roles")
plt.xlabel("Average Yearly Salary")
plt.ylabel("Density")

plt.xlim(0,400_000)
plt.tight_layout()
plt.show()


```
