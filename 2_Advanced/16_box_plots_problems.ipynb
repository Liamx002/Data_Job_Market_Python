```python
# importing data
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# importing dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
jobs = ["Data Analyst", "Data Engineer", "Data Scientist"]
df3 = df[(df["job_title_short"].isin(jobs)) & (df["job_country"] == "United Kingdom")].copy()
df3 = df3.dropna(subset = "salary_year_avg")
df3 = df3[df3["salary_year_avg"] > 0]
job_list = [df3[df3["job_title_short"] == job]["salary_year_avg"] for job in jobs]
plt.boxplot(job_list, tick_labels = jobs, vert = False)
plt.title("Hourly Pay for Each Job in the United Kingdom")
plt.xlabel("Average Yearly Salary")
plt.xlim(0,300_000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${int(x/1000)}K"))
plt.tight_layout()
plt.show()
```
