```python
# importing data
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# importing dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python

job_titles = ["Data Analyst", "Data Scientist", "Data Engineer"]
df_filtered = df[df["job_title_short"].isin(job_titles)]

df_filtered = df_filtered.groupby(["job_title_short", "job_schedule_type"]).agg(job_count=("job_title_short", "size"), salary_median=("salary_year_avg", "median")).fillna(0).reset_index()
df_filtered = df_filtered.sort_values(by="job_count", ascending=False)
df_filtered
```

```python
fig, ax = plt.subplots(1, 3)
job_number = 5
for i, job in enumerate(job_titles):
    skills_count = df_filtered[df_filtered["job_title_short"] == job].head(job_number)
    skills_count.plot(kind="scatter", x = "job_count", y = "salary_median", ax=ax[i])

    for j in range(len(skills_count)):
        ax[i].text(skills_count["job_count"].iloc[j],
                   skills_count["salary_median"].iloc[j],
                   skills_count["job_schedule_type"].iloc[j],
                   fontsize=9)
        
plt.tight_layout()

# Jesus my head hurts! :o
```
