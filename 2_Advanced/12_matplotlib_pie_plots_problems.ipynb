```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
fig, ax = plt.subplots(1, 3)

metrics = {
    "job_health_insurance": "Health Insurance Prov.",
    "job_schedule_type": "What Type Job Sched.",
    "job_no_degree_mention": "Degree Req."
}

for i, (column,name) in enumerate(metrics.items()):
    ax[i].pie(df[column].value_counts().head(2), startangle=90, autopct="%1.1f%%", labels=["False", "True"])
    ax[i].set_title(name)

fig.tight_layout()
plt.show()
# Keep in mind that the what type job schedule has shoddy true false format to accomodate the labels put in place ;(
```
