```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
df_job_country_salary = df.pivot_table(values="salary_year_avg", index="job_country", columns="job_title_short", aggfunc="median")
```

```python
top_countries = df["job_country"].value_counts().sort_values(ascending=False).head(6).index

df_top_countries = df_job_country_salary.loc[top_countries]
job_roles = ["Data Analyst", "Data Engineer", "Data Scientist"]
df_top_countries = df_top_countries[job_roles]
df_top_countries.plot(kind="bar")
plt.xticks(rotation=45)
plt.ylabel("Median Salary ($USD)")
plt.title("Median Salary by Country and Job Title")
plt.show()
```
