```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
df_1 = df.copy()
```

```python
df_1["job_posted_month"] = df_1["job_posted_date"].dt.strftime("%b")
df_1
```

```python
months = df_1["job_posted_month"].unique()

dict_months = {month:df_1[df_1["job_posted_month"] == month] for month in months}
dict_months["Jan"]
```

```python
df_q1 = pd.concat([dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]], ignore_index=True)
```

```python
df_q1["job_posted_month"].value_counts().plot(kind="bar")
```

```python
df_2 = df.copy()
df_2["job_posted_month"] = df_2["job_posted_date"].dt.strftime("%b")
```

```python
months = df_2["job_posted_month"].unique()
dict_months_2 = {month:df_2[df_2["job_posted_month"] == month] for month in months}
```

```python
comb_q2 = pd.concat([dict_months_2["Apr"], dict_months_2["May"], dict_months_2["Jun"]])
comb_q2["job_posted_month"].value_counts().plot(kind="bar")
```
